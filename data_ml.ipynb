```python
teams_data = pd.read_csv('./data/teams_data.csv')
```

```python
teams_data
```

```python
df=teams_data.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Selecting columns for normalization
columns_to_normalize = ['Rating', 'Average Combat Score', 'Kills:Deaths', 'Kill, Assist, Trade, Survive %',
                        'Average Damage Per Round', 'Kills Per Round', 'Assists Per Round',
                        'First Kills Per Round', 'First Deaths Per Round']

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

# Weight Assignment
weights = {
    'Rating': 0.2,
    'Average Combat Score': 0.2,
    'Kills:Deaths': 0.15,
    'Kill, Assist, Trade, Survive %': 0.1,
    'Average Damage Per Round': 0.1,
    'Kills Per Round': 0.1,
    'Assists Per Round': 0.05,
    'First Kills Per Round': 0.05,
    'First Deaths Per Round': -0.025
}

# Composite Score Calculation
df['Composite Score'] = (
    df['Rating'] * weights['Rating'] +
    df['Average Combat Score'] * weights['Average Combat Score'] +
    df['Kills:Deaths'] * weights['Kills:Deaths'] +
    df['Kill, Assist, Trade, Survive %'] * weights['Kill, Assist, Trade, Survive %'] +
    df['Average Damage Per Round'] * weights['Average Damage Per Round'] +
    df['Kills Per Round'] * weights['Kills Per Round'] +
    df['Assists Per Round'] * weights['Assists Per Round'] +
    df['First Kills Per Round'] * weights['First Kills Per Round'] +
    df['First Deaths Per Round'] * weights['First Deaths Per Round']
)

# Ranking
df['Rank'] = df['Composite Score'].rank(ascending=False)

# Display the ranked teams
df = df.sort_values(by='Rank')

df[['Teams', 'Rank', 'Composite Score']]
```
