```python
import numpy as np
import pandas as pd
```

```python
matches_scores= pd.read_csv('./vct_2024/matches/scores.csv')
```

```python
matches_scores
```

```python
matches_scores_AB=matches_scores[['Team A','Team B','Team A Score','Team B Score','Match Result']]
```

```python
matches_scores_AB.columns=['Team_A','Team_B','Team_A_Score','Team_B_Score','Match_Result']
```

```python
matches_scores_AB['Match_Result']=matches_scores_AB['Match_Result'].apply(lambda x: x.strip('won').strip())
```

```python
matches_scores_AB.iloc[0]['Match_Result']
```

```python
matches_scores_AB.Match_Result.value_counts()
```

```python
matches_scores_AB[['Team_A','Team_B','Team_A_Score','Team_B_Score']]
```

```python
df=matches_scores_AB[['Team_A','Team_B','Team_A_Score','Team_B_Score']].copy()
scores = {}

# Function to update scores
def update_scores(team, score):
    if team in scores:
        scores[team].append(score)
    else:
        scores[team] = [score]

# Update scores from the DataFrame
for _, row in df.iterrows():
    update_scores(row['Team_A'], row['Team_A_Score'])
    update_scores(row['Team_B'], row['Team_B_Score'])

# Convert the scores dictionary to a DataFrame
transposed_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in scores.items()]))
transposed_df.fillna(0, inplace=True)  # Optional: fill NaN values with 0

# Display the transposed DataFrame
print(transposed_df)
```

```python
transposed_df.mean().sort_values(ascending=False)
```

```python
teams_qualified = [
    "DRX",
    "Gen.G",
    "Paper Rex",
    "Talon Esports",
    "EDward Gaming",
    "FunPlus Phoenix",
    "Trace Esports",
    "Bilibili Gaming",
    "FNATIC",
    "FUT Esports",
    "Team Heretics",
    "Team Vitality",
    "Leviatán",
    "Sentinels",
    "G2 Esports",
    "KRÜ Esports"
]
```

```python
transposed_df.columns
```

```python
filtered_df = transposed_df.loc[:, transposed_df.columns.isin(teams_qualified)]
```

```python
filtered_df.head()
```

```python
filtered_df1= filtered_df.mean().reset_index()
```

```python
matches_res = matches_scores_AB.Match_Result.value_counts()
```

```python
filtered_df2 = matches_res[matches_res.index.isin(teams_qualified)].reset_index()
```

```python
filtered_df1.columns=['team','mean_win']
```

```python
filtered_df2.columns=['team','count_win']
```

```python
matches_win=filtered_df1.merge(filtered_df2,on='team')
```

```python
matches_win.to_csv('./data/matches_win_16.csv',index=False)
```

```python
player_data=['Player','Teams','Rounds Played', 'Rating', 'Average Combat Score', 'Kills:Deaths',
       'Kill, Assist, Trade, Survive %', 'Average Damage Per Round',
       'Kills Per Round', 'Assists Per Round', 'First Kills Per Round',
       'First Deaths Per Round','Maximum Kills in a Single Map', 'Kills',
       'Deaths', 'Assists', 'First Kills', 'First Deaths']
```

```python
players_df=players[player_data]
```

```python
players_df=players_df[players_df.Teams.isin(teams_qualified)]
```

```python
players_df['Kill, Assist, Trade, Survive %']=players_df['Kill, Assist, Trade, Survive %'].str.strip('%').astype(float)
```

```python
agg_dict={'Rounds Played':'sum', 'Rating':'mean', 'Average Combat Score':'mean',
       'Kills:Deaths':'mean', 'Kill, Assist, Trade, Survive %':'mean',
       'Average Damage Per Round':'mean', 'Kills Per Round':'sum', 'Assists Per Round':'sum',
       'First Kills Per Round':'sum', 'First Deaths Per Round':'sum',
       'Maximum Kills in a Single Map':'sum', 'Kills':'sum', 'Deaths':'sum', 'Assists':'sum',
       'First Kills':'sum', 'First Deaths':'sum'}
```

```python
teams_st=players_df.groupby('Teams',as_index=False).agg(agg_dict)
```
